```python
import pandas as pd
from time import sleep
from tqdm import tqdm

from selenium import webdriver
from selenium.common.exceptions import WebDriverException
```

```python
products = pd.read_csv('Products.tsv', sep='\t')
products
```

```python
driverPath = './chromedriver'
```

```python
driver = webdriver.Chrome(executable_path=driverPath)
# options = webdriver.ChromeOptions()
# options.add_argument('window-size=1200x1000')
```

```python
X_Paths = {
    'ProductRating': '/html/body/div[7]/article/div[5]/div/div[2]/section/div[2]/div/div[1]/div/div[1]/div/div/div[1]',
    'Brand': '/html/body/div[7]/article/div[1]/div[2]/section[3]/div/div[3]/a/span/bdi',
    'RatingCount': '/html/body/div[7]/article/div[5]/div/div[2]/section/div[2]/div/div[1]/div/div[1]/div/div/div[2]/span',
    'Q&ACount': '/html/body/div[7]/article/div[1]/div[2]/section[3]/div/div[4]/span[2]/a/span',
    'ProductSpecList': '/html/body/div[7]/article/div[1]/div[2]/section[5]/ul',
    'ProductDescription': '/html/body/div[7]/article/div[2]/div/section/div[2]'
}
```

```python
def getTextfromXpath(driver, xPath):
    try:
        element = driver.find_element_by_xpath(xPath)
    except:
        return None
    return element.text
```

```python
def getInfofromPage(driver):
    result = {}

    for info, xPath in X_Paths.items():
        result[info] = getTextfromXpath(driver, xPath)
        
    return result
```

```python
# get information
all_info = []

for url in tqdm(products.product_url):
    # load the webpage
    driver.get(url)
    # sleep for loading
    sleep(2)
    
    info = getInfofromPage(driver)
    all_info.append(info)
```

```python
extra_info = pd.DataFrame(all_info)
```

```python
extra_info[['product_id','product_url']] = products[['product_id', 'product_url']]
```
